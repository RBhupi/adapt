```python

import sqlite3
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

DB_PATH = "/Users/bhupendra/projects/arm_radar_scaning/output/adapt-output-bnf-khtx-20250617/analysis/KHTX_cells_statistics.db"
```

```python


def connect_db(path):
    return sqlite3.connect(path)

def get_columns(conn, table):
    q = f"PRAGMA table_info({table})"
    return pd.read_sql(q, conn)['name'].to_list()

def fetch(conn, table, cols, condition):
    cols = ", ".join(cols)
    if not condition.strip():
        query = f"SELECT {cols} FROM {table}"
    else:
        query = f"SELECT {cols} FROM {table} WHERE ({condition})"
    return pd.read_sql(query, conn)

def build_selector(cols):
    return widgets.SelectMultiple(options=cols, value=tuple(cols), description="cols")
```

```python

def build_condition_box(value):
    return widgets.Text(
        value=value,
        description="where",
        placeholder="SQL condition"
    )

def run_query(_):
    chosen = list(selector.value)
    cond = condition_box.value
    df = fetch(conn, "cells", chosen, cond)
    display(df)
```

```python

conn = connect_db(DB_PATH)
columns = get_columns(conn, "cells")

selector = build_selector(columns)
condition_box = build_condition_box(value="cell_area_sqkm > 300 AND radar_reflectivity_mean > 40 AND radar_differential_reflectivity_max < 2")
button = widgets.Button(description="Show")
button.on_click(run_query)

display(selector, condition_box, button)
```
